```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.info()
test.info()
```

```python
def compute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):

        if Pclass == 1:
            return 39

        elif Pclass == 2:
            return 32.5

        else:
            return 28.25

    else:
        return Age
```

```python
train['Age'] = train[['Age','Pclass']].apply(compute_age, axis = 1)
test['Age'] = test[['Age', 'Pclass']].apply(compute_age, axis = 1)
train
```

```python
def compute_cabins(cols):
    cabin = cols[0]
    
    if pd.isnull(cabin):
        return 'O'
    
    else:
        list1 = list()
        for char in cabin:
            list1.append(char)
        return list1[0]
```

```python
train['Cabin'] = train[['Cabin']].apply(compute_cabins, axis = 1)
test['Cabin'] = test[['Cabin']].apply(compute_cabins, axis = 1)
train.head()
```

```python
def compute_bcde(cols):
    cabin = cols[0]
    
    if cabin == 'B' or cabin == 'C' or cabin == 'D' or cabin == 'E':
        return 1
    else:
        return 0
```

```python
train['bcde'] = train[['Cabin']].apply(compute_bcde, axis = 1)
test['bcde'] = test[['Cabin']].apply(compute_bcde, axis = 1)
train
```

```python
def compute_ticket_type(cols):
    ticket = cols[0]
    ticket = str(ticket)
    
    if ((ticket[0] >= 'a' and ticket[0] <= 'z') or (ticket[0] >= 'A' and ticket[0] <= 'Z')):
        return ticket[0]
    else:
        return 'other'
    
```

```python
train['ticket_type'] = train[['Ticket']].apply(compute_ticket_type, axis = 1)
test['ticket_type'] = test[['Ticket']].apply(compute_ticket_type, axis = 1)
train
```

```python
def compute_ticket_len(cols):
    ticket = cols[0]
    ticket = str(ticket)
    
    return len(ticket)
```

```python
train['ticket_len'] = train[['Ticket']].apply(compute_ticket_len, axis = 1)
test['ticket_len'] = test[['Ticket']].apply(compute_ticket_len, axis = 1)
train
```

```python
def compute_sex(cols):
    sex = cols[0]
    age = cols[1]
    
    if age <= 16 :
        return 'child'
    else :
        return sex
```

```python
train['sex2'] = train[['Sex', 'Age']].apply(compute_sex, axis = 1)
test['sex2'] = test[['Sex', 'Age']].apply(compute_sex, axis = 1)
train
```

```python
def age_division(cols):
    age = cols[0]
    
    if age<=18 :
        return 'small'
    elif age > 18 and age <= 35:
        return 'medium'
    else:
        return 'large'
```

```python
train['age2'] = train[['Age']].apply(age_division, axis = 1)
test['age2'] = test[['Age']].apply(age_division, axis = 1)
test.head()
```

```python
train.info()
test.info()
```

```python
train.dropna(inplace = True)
test.fillna(test.mean(),inplace = True)
train.info()
test.info()
```

```python
import re
train['u_name'] = train['Name'].transform(lambda f:re.findall('(Mrs|Mr|Master|Miss|Rev.)',f))
test['u_name'] = test['Name'].transform(lambda f:re.findall('(Mrs|Mr|Master|Miss|Rev.)',f))
train.info()
```

```python
def compute_names(cols):
    names = cols[0]
    
    if len(names)==0:
        return 'other'
    
    else:
        return names[0]
```

```python
train['u_name'] = train[['u_name']].apply(compute_names, axis = 1)
test['u_name'] = test[['u_name']].apply(compute_names, axis = 1)
train
```

```python
train.info()
test.info()
```

```python
family = pd.DataFrame(train['SibSp']+train['Parch'], columns =['family'] )
train = pd.concat([train,family], axis = 1)
family2 = pd.DataFrame(test['SibSp']+test['Parch'], columns = ['family'])
test = pd.concat([test, family2], axis = 1)
train.info()
test.info()
```

```python
def compute_alone(cols):
    alone = cols[0]
    
    if alone == 0:
        return 1
    
    else:
        return 0
```

```python
train['alone'] = train[['family']].apply(compute_alone, axis = 1)
test['alone'] = test[['family']].apply(compute_alone, axis = 1)
train.head()
```

```python
def compute_namelen(cols):
    name = cols[0]
    
    
    return len(name)
```

```python
train['name_len'] = train[['Name']].apply(compute_namelen, axis = 1)
test['name_len'] = train[['Name' ]].apply(compute_namelen, axis = 1)
train.info()
test.info()
```

```python
def compute_faretype(cols):
    fare = cols[0]
    
    if fare >= 80:
        return 1
    else:
        return 0
    
```

```python
train['fare_above_80'] = train[['Fare']].apply(compute_faretype, axis = 1)
test['fare_above_80'] = train[['Fare' ]].apply(compute_faretype, axis = 1)
train.info()
test.info()
```

```python
def compute_familytype(cols):
    family = cols[0]
    
    if family < 1:
        return 'small'
    elif family >= 1 or family <= 3:
        return 'medium'
    else:
        return 'large'
    
```

```python
train['family_type'] = train[['family']].apply(compute_familytype, axis = 1)
test['family_type'] = train[['family' ]].apply(compute_familytype, axis = 1)
train.info()
test.info()
```

```python
test['family_type'].fillna(test['family_type'].mode()[0], inplace = True)
train.info()
test.info()
```

```python
train['age2_sex'] = train['age2']+train['Sex']
test['age2_sex'] = test['age2']+test['Sex']
train.info()
test.info()
```
