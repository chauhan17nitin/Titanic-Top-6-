```python
def compute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):

        if Pclass == 1:
            return 39

        elif Pclass == 2:
            return 32.5

        else:
            return 28.25

    else:
        return Age
```

```python
train['Age'] = train[['Age','Pclass']].apply(compute_age, axis = 1)
test['Age'] = test[['Age', 'Pclass']].apply(compute_age, axis = 1)
train
```

```python
def compute_cabins(cols):
    cabin = cols[0]
    
    if pd.isnull(cabin):
        return 'O'
    
    else:
        list1 = list()
        for char in cabin:
            list1.append(char)
        return list1[0]
```

```python
train['Cabin'] = train[['Cabin']].apply(compute_cabins, axis = 1)
test['Cabin'] = test[['Cabin']].apply(compute_cabins, axis = 1)
train.head()
```

```python
def compute_ticket_type(cols):
    ticket = cols[0]
    ticket = str(ticket)
    
    if ((ticket[0] >= 'a' and ticket[0] <= 'z') or (ticket[0] >= 'A' and ticket[0] <= 'Z')):
        return ticket[0]
    else:
        return 'other'
    
```

```python
train['ticket_type'] = train[['Ticket']].apply(compute_ticket_type, axis = 1)
test['ticket_type'] = test[['Ticket']].apply(compute_ticket_type, axis = 1)
train
```

```python
def compute_ticks(cols):
    ticket_len = cols[0]
    ticket_type = cols[1]
    
    return str(ticket_len)+str(ticket_type)
```

```python
train['type_len'] = train[['ticket_lentype', 'ticket_type']].apply(compute_ticks, axis = 1)
test['type_len'] = test[['ticket_lentype', 'ticket_type']].apply(compute_ticks, axis = 1)
train
```

```python
def compute_sex(cols):
    sex = cols[0]
    age = cols[1]
    
    if age <= 16 :
        return 'child'
    else :
        return sex
```

```python
def age_division(cols):
    age = cols[0]
    
    if age<=18 :
        return 'small'
    elif age > 18 and age <= 35:
        return 'medium'
    else:
        return 'large'
```

```python
train['age2'] = train[['Age']].apply(age_division, axis = 1)
test['age2'] = test[['Age']].apply(age_division, axis = 1)
test
```

```python
train['sex2'] = train[['Sex', 'Age']].apply(compute_sex, axis = 1)
test['sex2'] = test[['Sex', 'Age']].apply(compute_sex, axis = 1)
train
```
